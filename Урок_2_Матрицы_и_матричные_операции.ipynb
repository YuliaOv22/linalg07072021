**__5*.__ Написать на Python функцию для перемножения двух произвольных матриц, не используя NumPy.**

```python
def matrixdot(matrix_1, matrix_2):
  
  #Заполняем будущую матрицу нулями
  result_matrix = [[0 for i in range(len(matrix_2[0]))] for i in range(len(matrix_1))]
  
  # Проверяем размерности входящих матриц
  if len(matrix_1[0]) == len(matrix_2):

    # Находим произведение входящих матриц
    for i in range(0, len(matrix_1)):
      for j in range(0, len(matrix_2[0])):
        result_matrix[i][j] = sum(matrix_1[i][k] * matrix_2[k][j] for k in range(len(matrix_2)))
    
    return result_matrix

  else:
    return "Размерности матриц не совпадают!"
```

```python
# Решение

def skalar_prod(v1, v2):
    return sum([i * j for i, j in zip(v1, v2)])
       
def matrix_product(A, B, display_result=True):
    result = []
    try:
        for row_a in A:
            internal_row = []
            for k in range(len(B[0])):
                column = [B[i][k] for i in range(len(B))]
                internal_row.append(skalar_prod(row_a, column))
            result.append(internal_row)
        if display_result:
            for row in result:
                print(row)
        return result
    except Exception as e:
        print('wrong dimensions!')
        print(e)
```

**__2.__ Определитель матрицы $A$ равен $4$. Найти:**

   а) $det(A^{2})$;

   в) $det(2A)$.

$det(A^{2}) = detA * detA = 4 * 4 = {\color {red}{16}}$
    
   **б)** Воспользуемся свойством определителя. Определитель транспонированной матрицы равен определителю исходной: $detA^{T}=detA$.

   Так как порядок матрицы не задан, получаем:
   
   $det(2A) = {\color {red}{2^n * 4}}$.

```python
import numpy as np
import scipy.linalg

A = np.array([[2, 0], [0, 2]])
A_ = np.array([[2, 0, 0], [0, 2, 0] , [0, 0, 1]])

print(f'Определитель матрицы A:\n{np.linalg.det(A):.0f}')
print(f'Определитель матрицы A^2:\n{np.linalg.det(A*A):.0f}')
print(f'Определитель матрицы A^T:\n{np.linalg.det(A.T):.0f}')
print(f'Определитель матрицы 2-го порядка 2A:\n{np.linalg.det(2*A):.0f}')
print(f'Определитель матрицы 3-го порядка 2A_:\n{np.linalg.det(2*A_):.0f}')
```

**__4.__ Найти ранг матрицы:**

   а) $\begin{pmatrix}
1 & 2 & 3\\ 
1 & 1 & 1\\ 
2 & 3 & 4
\end{pmatrix};$

   б) $\begin{pmatrix}
0 & 0 & 2 & 1\\ 
0 & 0 & 2 & 2\\

\end{pmatrix}.$

   **а)** Ранг матрицы равен ${\color {red}{2}}$.
   
   $\begin{pmatrix}

   **б)** Ранг матрицы равен ${\color {red}{3}}$.
   
   $\begin{pmatrix}

```python
import numpy as np
import scipy.linalg

A = np.array([[1, 2, 3], [1, 1, 1], [2, 3, 4]])
B = np.array([[0, 0, 2, 1], [0, 0, 2, 2], [0, 0, 4, 3], [2, 3, 5, 6]])
rA = np.linalg.matrix_rank(A)
rB = np.linalg.matrix_rank(B)

print(f'Ранг матрицы A: {rA}')
print(f'Ранг матрицы B: {rB}')
```
