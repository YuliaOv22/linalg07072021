**__1.__ Решить систему уравнений методом Гаусса:**

$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
2x_{1}+x_{2}-x_{3}+x_{4}=-2, \\
x_{1}+x_{2}-3x_{3}+x_{4}=4.
\end{cases}$

Запишем матрицу коэффициентов и преобразуем ее к треугольному виду:

$A = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 & -2 \\ 
2 & 1 & -1 & 1 \\ 
1 & 1 & -3 & 1
\end{matrix}\right|
\begin{matrix}
0\\ 
-2\\
4
\end{matrix}
\end{pmatrix}\Rightarrow \begin{pmatrix}
\left.\begin{matrix}
2 & 1 & -1 & 1 \\ 
0 & -1 & 1 & 5 \\ 
0 & 0 & -4 & 6
\end{matrix}\right|
\begin{matrix}
-2\\ 
-2\\
8
\end{matrix}
\end{pmatrix}$

Получим следующую систему уравнений:

$\begin{cases}
2x_{1}+x_{2}-x_{3}+x_{4}=-2\\
-x_{2}+x_{3}+5x_{4}=-2\\
-4x_{3}+6x_{4}=8
\end{cases}$

Так как в системе осталось $2$ неизвестных, значит, она имеет бесконечное число решений. Обозначим $x_{4} = c$ и выразим остальные неизвестные через нее.

${\color {red}{\begin{cases}
x_{4} = c\\
x_{3}=-\frac{4-3c}{2}\\
x_{2}=2+\frac{7c-4}{2}\\
x_{1}=-2-3c\\
\end{cases}}}$

Таким образом, мы нашли **общее решение** системы.

**Частным решением** системы будет, если, к примеру, мы возьмем $c=0$, тогда:

${\color {red}{\begin{cases}
x_{4} = 0\\
x_{3}=-2\\
x_{2}=1\\
x_{1}=-2\\
\end{cases}}}$

   **а)** $\begin{cases}

Так как количество уравнений совпадает с количеством неизвестных, правило Крамера можно применить. Найдем определитель матрицы коэффициентов:

$detA=\begin{vmatrix}
1 & -2\\ 
3 & -4
\end{vmatrix}=1*(-4)-3*(-2)=-4+6=2\neq 0$

Определитель не равен нулю, следовательно, система совместна.

Найдем определители $detA_{1}$, $detA_{2}$:

$detA_{1}=\begin{vmatrix}
1 & -2\\ 
7 & -4
\end{vmatrix}=1*(-4)-7*(-2)=-4+14=10$

$detA_{2}=\begin{vmatrix}
1 & 1\\ 
3 & 7
\end{vmatrix}=1*7-3*1=7-3=4$

Найдем решение по формулам Крамера:

$x_{1} = \frac{detA_{1}}{detA} = \frac{10}{2}={\color {red}{5}}$

$x_{2} = \frac{detA_{2}}{detA} = \frac{4}{2}={\color {red}{2}}$

```python
import numpy as np

Aa = np.array([[1, -2], [3, -4]])
ba = np.array([1, 7])
detAa = np.linalg.det(Aa)
Aa1 = np.column_stack([ba, Aa.T[1]])
Aa2 = np.column_stack([Aa.T[0], ba])

print(f'a) \
      \nКоличество уравнений: {Aa.shape[0]}, \
      \nКоличество неизвестных: {Aa.shape[1]}, \
      \nОпределелитель матрицы А: {detAa:.0f}, \
      \nОпределелитель матрицы А1: {np.linalg.det(Aa1):.0f}, \
      \nОпределелитель матрицы А2: {np.linalg.det(Aa2):.0f} \
      \nx1: {np.linalg.det(Aa1)/detAa:.0f} \
      \nx2: {np.linalg.det(Aa2)/detAa:.0f}')
```

**__3*.__ Решить систему линейных уравнений методом $LU$-разложения**

$\begin{cases}
2x_{1}+x_{2}+3x_{3}=1 \\
11x_{1}+7x_{2}+5x_{3}=-6 \\
9x_{1}+8x_{2}+4x_{3}=-5
\end{cases}$

Матрица коэффициентов будет иметь вид:

$A=\begin{pmatrix}
2 & 1 & 3\\ 
11 & 7 & 5\\ 
9 & 8 & 4
\end{pmatrix}$

Найдем матрицы $U$ и $L$:

$U=\begin{pmatrix}
2 & 1 & 3\\ 
0 & \frac{3}{2} & -\frac{23}{2}\\ 
0 & 0 & \frac{52}{3}
\end{pmatrix}$

$L=\begin{pmatrix}
1 & 0 & 0\\ 
\frac{11}{2} & 1 & 0\\ 
\frac{9}{2} & \frac{7}{3} & 1
\end{pmatrix}$

Решим систему $Ly=b:$

$\begin{cases}
y_{1}=1 \\
\frac{11}{2}y_{1}+y_{2}=-6 \\
\frac{9}{2}y_{1}+\frac{7}{3}y_{2}+y_{3}=-5
\end{cases}\ \ \ \Rightarrow \ \ \ \begin{cases}
y_{1}=1\\
y_{2}=-\frac{23}{2}\\
y_{3}=\frac{52}{3}
\end{cases}$

Далее решим систему $Ux=y:$

$\begin{cases}
2x_{1}+x_{2}+3x_{3}=1 \\
\frac{3}{2}x_{2}-\frac{23}{2}x_{3}=-\frac{23}{2} \\
\frac{52}{3}x_{3}=\frac{52}{3}
\end{cases}\ \ \ \Rightarrow \ \ \ {\color {red}{\begin{cases}
x_{1}=-1 \\
x_{2}=0 \\
x_{3}=1
\end{cases}}}$

Произведем проверку, подставив полученные значения переменных в исходную систему:

$\begin{cases}
2*(-1)+0+3*1=1 \\
11*(-1)+7*0+5*1=-6 \\
9*(-1)+8*0+4*1=-5
\end{cases}\ \ \ \Rightarrow \ \ \ \begin{cases}
-2+3=1 \\
-11+5=-6 \\
-9+4=-5
\end{cases}$

Таким образом, найденное решение верно.

```python
import numpy as np
from scipy.linalg import lu_factor, lu_solve

A = np.array([[2, 1, 3], [11, 7, 5], [9, 8, 4]])
b = np.array([1, -6, -5])
lu, piv = lu_factor(A)
x = lu_solve((lu, piv), b)

print(f'Значения x: {x}')
print(f'Проверка Ax - b = 0: {np.allclose(A @ x - b, np.zeros((3,)))}')
```

**__4*.__ Решить систему линейных уравнений методом Холецкого**

$\begin{cases}
81x_{1}-45x_{2}+45x_{3}=531 \\
-45x_{1}+50x_{2}-15x_{3}=-460 \\
45x_{1}-15x_{2}+38x_{3}=193
\end{cases}$

Произведем разложение на $LL^{T}$:

$l_{11}=\sqrt{a_{11}}=\sqrt{81} = 9$

$l_{21}=\frac{a_{21}}{l_{11}}=-\frac{45}{9} = -5$

$l_{31}=\frac{a_{31}}{l_{11}}=\frac{45}{9} = 5$

$l_{22}=\sqrt{a_{22}-l_{21}^{2}}=\sqrt{50 - (-5)^2}=5$

$l_{32}=\frac{1}{l_{22}}\left ( a_{32}-l_{21}l_{31} \right)=\frac{1}{5}\left ( -15 - (-5)*5 \right )=2$

$l_{33}=\sqrt{a_{33}-l_{31}^{2}-l_{32}^{2}}=\sqrt{38-5^2-2^2}=3$

Получим матрицы:

$L = \begin{pmatrix}
9 & 0 & 0 \\ 
-5 & 5 & 0 \\ 
5 & 2 & 3
\end{pmatrix}, 
\; \; 
L^{T} = \begin{pmatrix}
9 & -5 & 5 \\ 
0 & 5 & 2 \\ 
0 & 0 & 3
\end{pmatrix}$

Решим систему $Ly=b:$

$\begin{cases}
9y_{1}=531 \\
-5y_{1}+5y_{2}=-460 \\
5y_{1}+2y_{2}+3y_{3}=193
\end{cases}\ \ \ \Rightarrow \ \ \ \begin{cases}
y_{1}=59 \\
y_{2}=-33 \\
y_{3}=-12
\end{cases}$

Далее решим систему $L^{T}x=y:$

$\begin{cases}
9x_{1}-5x_{2}+5x_{3}=59 \\
5x_{2}+2x_{3}=-33 \\
3x_{3}=-12
\end{cases}\ \ \ \Rightarrow \ \ \ \ {\color {red}{\begin{cases}
x_{1}=6 \\
x_{2}=-5 \\
x_{3}=-4
\end{cases}}}$

Осуществим проверку, подставив полученные значения в исходную систему:

$\begin{cases}
81*6-45*(-5)+45*(-4)=531 \\
-45*6+50*(-5)-15*(-4)=-460 \\
45*6-15*(-5)+38*(-4)=193
\end{cases}\ \ \ \Rightarrow \ \ \ \begin{cases}
486+225-180=531 \\
-270-250+60=-460 \\
270+75-152=193
\end{cases}$

```python
import numpy as np
from scipy.linalg import cho_factor, cho_solve

A = np.array([[81, -45, 45], [-45, 50, -15], [45, -15, 38]])
b = np.array([531, -460, 193])
c, low = cho_factor(A)
x = cho_solve((c, low), b)

print(f'Значения x: {x}')
print(f'Проверка Ax - b = 0: {np.allclose(A @ x - b, np.zeros((3,)))}')
```
